```````python
import requests

url = 'http://localhost:11434/api/generate'

data = {
    "model": "gemma3:4b",
    "property": "天空為什麼是藍的?"
}

response = requests.post(url, json= data)

if response.status_code == 200:
    print("下載成功!")
    print("回傳的")
else:
    print("下載失敗")
print(type(response))
```````

```````python
import requests
from IPython.display import display, Markdown

def chat_with_ollama(prompt: str):
    url = "http://localhost:11434/api/generate"
    payload = {
        "model": "gemma3:4b",
        "prompt": prompt,
        "stream": False,
        "options": {  # 生成設定
            "temperature": 0.7,
            "top_p": 0.9,
            "top_k": 50,
        },

    }

    response = requests.post(url, json=payload)
    result = response.json()

    # 偵測可能的回應 key
    if "response" in result:
        text = result["response"]
    elif "message" in result:
        text = result["message"]
    elif "content" in result:
        text = result["content"]
    else:
        text = f"⚠️ No expected key found.\n\nAvailable keys: {list(result.keys())}\n\n``````"

    # 使用 Markdown 在 Jupyter inline 顯示
    display(Markdown(f"### 💬 AI 回應：\n\n{text}"))

# 範例輸入
chat_with_ollama("請用簡單的方式解釋什麼是Python的函式？")
```````
